```python
import csv
import json
import random
from tqdm import tqdm
from collections import defaultdict

"""Data preprocess for FlashRAG ODQA datasets
Data Link: https://huggingface.co/datasets/RUC-NLPIR/FlashRAG_datasets
"""

dataset_name = 'hotpotqa'
split = 'train'
# dataset_name = '2wiki'
# split = 'dev'
sample_num = 500  # Renamed data_num to sample_num for clarity

test_path = f'/workspace/Search-o1/data/FlashRAG_datasets/{dataset_name}/{split}.jsonl'
output_path = f'./QA_Datasets/{dataset_name}_with_ctx.json'

all_lines = []
# First, read all lines from the file
print(f"Reading all lines from {test_path}...")
with open(test_path, 'r', encoding='utf-8') as file:
    for line in file:
        all_lines.append(line)

print(f"Total lines read: {len(all_lines)}")

# Determine how many samples to take
if len(all_lines) <= sample_num:
    print(f"Total lines ({len(all_lines)}) are less than or equal to sample_num ({sample_num}). Using all available lines.")
    sampled_lines = all_lines
else:
    print(f"Randomly sampling {sample_num} lines from {len(all_lines)} total lines...")
    sampled_lines = random.sample(all_lines, sample_num)

data_list = []
# Process the sampled lines
for idx, line_str in enumerate(tqdm(sampled_lines, desc="Processing sampled data")):
    # try:
        line = json.loads(line_str)


        titles = line['metadata']['context']['title']
        contents = line['metadata']['context']['content']
        gold_titles = line['metadata']['supporting_facts']['title']
        gold_sent_ids = line['metadata']['supporting_facts']['sent_id']
        
        context_by_title = {}

        for title, sent_id in zip(gold_titles, gold_sent_ids):
            try:
                title_idx = titles.index(title)
                sentence = contents[title_idx][sent_id]
                context_by_title[title] = context_by_title.get(title, "") + " " + sentence
            except (ValueError, IndexError):
                print(f"警告：找不到 title '{title}' 或句子索引 {sent_id} 越界")
                continue

        paragraph_list = [f"{title}\n{text}" for title, text in context_by_title.items()]


        data_list.append({
            'id': idx, # Use a new sequential ID for the sampled data
            'Question': line['question'],
            'answer': line["golden_answers"],
            'golden_context': paragraph_list,
        })
    # except json.JSONDecodeError as e:
    #     print(f"Error decoding JSON on line {idx}: {line_str.strip()} - {e}")
    # except KeyError as e:
    #     print(f"Missing expected key on line {idx}: {line_str.strip()} - {e}")


# Write the updated data to JSON
print(f"Writing {len(data_list)} items to {output_path}...")
with open(output_path, mode='w', encoding='utf-8') as json_file:
    json.dump(data_list, json_file, indent=4, ensure_ascii=False)

print("Data preprocessing complete!")
```
